```python
import pandas as pd
import requests
import bs4
```

```python
# make the http request 
response = requests.get('https://web-scraping-demo.zgulde.net/people')
html = response.text

#turn the response into a beautiful soup object
soup = bs4.BeautifulSoup(response.text)
soup
```

```python
# find the container for the information we want
people_container = soup.select('#people')[0]
```

```python
people = people_container.select('.person')
```

```python
person = people[0]
```

```python
print(person.prettify())
```

```python
person.find(class_='name').text
person.find(class_='quote').text.strip()  #removing excess white space
person.find(class_='email').text
person.find(class_='phone').text
person.find(class_='address').text.strip() #removing excess white space
```

```python
#turn into a dictionary
def process_person(person):
    return {
        'name': person.find(class_='name').text,
        'quote': person.find(class_='quote').text.strip(),
        'email': person.find(class_='email').text,
        'phone': person.find(class_='phone').text,
        'address': person.find(class_='address').text.strip(),
    }
```

```python
process_person(people[9])
```

```python
pd.DataFrame([process_person(person) for person in people])
```
